```python
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.lsa import LsaSummarizer

# Example long text documents
long_text_1 = """
The quick brown fox jumps over the lazy dog. This classic sentence is a pangram, a sentence containing
every letter of the alphabet at least once. Pangrams are used in the study of typefaces and keyboard layouts,
and are also used to test fonts and keyboards to ensure they contain all the necessary characters. The
sentence was popularized in the 19th century, and has since become a common example of a pangram used in
various contexts, including educational materials and design samples.
"""

long_text_2 = """
Artificial intelligence (AI) refers to the simulation of human intelligence in machines that are programmed
to think and learn like humans. These intelligent machines can perform tasks that typically require human
cognition, such as understanding natural language, recognizing patterns, and making decisions. AI technologies
are rapidly advancing and are being used in various applications, including virtual assistants, autonomous
vehicles, and predictive analytics. The goal of AI research is to create systems that can solve complex problems
and improve the efficiency of processes across different industries.
"""

# Create parsers for the long text documents
parser_1 = PlaintextParser.from_string(long_text_1, Tokenizer("english"))
parser_2 = PlaintextParser.from_string(long_text_2, Tokenizer("english"))

# Create an LSA summarizer
summarizer = LsaSummarizer()

# Get summaries
summary_1 = summarizer(parser_1.document, 3)  # Get 3 sentences as summary
summary_2 = summarizer(parser_2.document, 4)  # Get 3 sentences as summary

# Print the summaries
print("Summary 1:", ' '.join([str(sentence) for sentence in summary_1]))
print("Summary 2:", ' '.join([str(sentence) for sentence in summary_2]))
```
